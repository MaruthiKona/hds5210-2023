**DATA MERGING-1**

 An inner merge between two datasets, `df2` and `df1`, utilizing common columns such as 'Country' and 'WHO Region.' This merging process is crucial for creating a unified dataset that combines pertinent information from both sources. By selecting specific columns from each dataset, such as the reported number of people receiving Antiretroviral Therapy (ART), estimated numbers of individuals living with HIV, and the ART coverage percentages, the resulting DataFrame, named `merged_data`, offers a consolidated view of key HIV/AIDS metrics. This approach facilitates a more comprehensive analysis of the relationship between these metrics within distinct WHO regions, ensuring that the data aligns accurately across corresponding countries. The merged dataset serves as a valuable resource for gaining insights into the HIV/AIDS landscape, enabling researchers, policymakers, and analysts to make informed decisions and develop a nuanced understanding of the status of HIV/AIDS across different geographical regions. Ultimately, the merging process enhances the overall reliability and integrity of the data, providing a more holistic perspective on the global HIV/AIDS scenario.

```python
# Select only the desired columns from each DataFrame
selected_columns_df1 = df2[["Country", "Reported number of people receiving ART",
                            "Estimated number of people living with HIV",
                            "Estimated ART coverage among people living with HIV (%)", 'WHO Region']]
selected_columns_df2 = df1[['Country', 'Received Antiretrovirals',
                            'Needing antiretrovirals', 'Percentage Recieved','WHO Region']]

# Merge DataFrames based on the 'Country' and 'WHO Region' columns
merged_data = pd.merge(selected_columns_df1, selected_columns_df2, on=['Country', 'WHO Region'], how='inner')

# Replace 'No data' with 0 in the entire DataFrame
merged_data.replace('No data', 0, inplace=True)

# Convert 'Reported number of people receiving ART' to numeric, coercing errors to NaN
merged_data["Reported number of people receiving ART"] = pd.to_numeric(merged_data["Reported number of people receiving ART"], errors='coerce')
```

**DATA TRANSFORMATIONS**

1.**RENAMING OF COLUMN NAMES**

In the given code, specific columns within the "merged_data" DataFrame have been renamed using a dictionary called "column_mapping." The original column "Reported number of people receiving ART" is now labeled as "New_ART_Receiving_Column_Name," offering a more abstract and adaptable identifier. Similarly, the columns "Estimated number of people living with HIV" and "Estimated ART coverage among people living with HIV (%)" have been renamed to "New_Living_with_HIV_Column_Name" and "New_ART_Coverage_Column_Name," respectively, providing concise and standardized labels. The resulting DataFrame displays these modified column names, with each row representing a distinct country or region along with various metrics related to HIV, antiretroviral therapy, and regional classification.

```python
# Renaming the column names
column_mapping = {
    'Reported number of people receiving ART': 'New_ART_Receiving_Column_Name',
    'Estimated number of people living with HIV': 'New_Living_with_HIV_Column_Name',
    'Estimated ART coverage among people living with HIV (%)': 'New_ART_Coverage_Column_Name'
}

# Use the rename method to apply the changes
merged_data.rename(columns=column_mapping, inplace=True)
merged_data
```

**2. Changing Float to Integer**

The code converts specific columns in the pandas DataFrame `merge` to the nullable integer type ('Int64') while handling NaN values. For the "Estimated ART coverage among children (%)_max" column, the `pd.to_numeric()` function is used with the `errors='coerce'` parameter to convert values to numeric format and replace non-numeric or NaN entries with NaN. Subsequently, the `astype('Int64')` method is applied to enforce the nullable integer type. The same approach is employed for the "Percentage Received_max" column. The resulting DataFrame, displayed using `merge.info()`, confirms the successful conversion of these columns to nullable integers, allowing for numeric operations while accommodating potential missing values.

```python
# Convert the column to int, handling NaN values
merge["Estimated ART coverage among children (%)_max"] = pd.to_numeric(merge["Estimated ART coverage among children (%)_max"], errors='coerce').astype('Int64')

# Convert the column to int, handling NaN values
merge["Percentage Recieved_max"] = pd.to_numeric(merge["Percentage Recieved_max"], errors='coerce').astype('Int64')
merge.info()
```

**4.MATHEMATICAL CALCUATION** & **ADDING THE NEW COLUMN**

The presented code extends the analytical capacity of the pre-existing DataFrame, "merge," by introducing a novel column denominated as "total_ART%." This augmentation is achieved through a mathematical operation involving two extant columns within the DataFrame, namely "Percentage Recieved_max" and "Estimated ART coverage among children (%)_max." The essence of this computation lies in the summation of values from these two columns, thereby yielding an integrated metric representative of the total antiretroviral therapy (ART) coverage percentage. This composite indicator serves to amalgamate information pertaining to both the reported percentage of individuals undergoing antiretroviral treatment and the estimated ART coverage among children. The resultant DataFrame manifests the original columns alongside the newly incorporated "total_ART%" column, which collectively provides a nuanced depiction of ART coverage. Importantly, the code exhibits due diligence in retaining NaN and "<NA>" entries from the contributing columns, thus maintaining the integrity of the dataset in the context of potential missing or undefined values. This calculated metric, "total_ART%," contributes substantively to a holistic understanding of the overarching ART coverage dynamics encapsulated within the dataset.

```python
merge["total_ART%"] = (merge["Percentage Recieved_max"]+ merge["Estimated ART coverage among children (%)_max"])

merge
```

**PIVOT 2**

The pivot table that was produced, "merge_pivoted," derived from the original DataFrame "merge," serves to elucidate the comprehensive distribution of aggregated antiretroviral therapy (ART) coverage percentages across various countries and their respective World Health Organization (WHO) regions. The pivotal focus lies in the "total_ART%" metric, where each cell in the table encapsulates the cumulative ART coverage percentage for a particular country within its designated WHO region. This structured representation enables a systematic examination of the disparities in ART coverage across diverse geographic areas. The aggregation function, employing summation, consolidates the ART coverage values for each country, contributing to a holistic understanding of the collective ART landscape within distinct WHO regions. The pivot table, adeptly handling potential missing values by imputing zeros, thus provides a nuanced and organized portrayal conducive to comparative analyses and strategic insights into the regional distribution of total ART percentages across the dataset.

```python
merge_pivoted = merge.pivot_table(
    values=['total_ART%'],
    index='Country',
    columns='WHO Region',
    aggfunc=np.sum,
    fill_value=0
)
merge_pivoted
```
